```python
print("Фильтрация данных")

data = [
    ["Али", 19, 85],
    ["Мария", 21, 92],
    ["Иван", 20, 78],
    ["Елена", 22, 95],
    ["Дмитрий", 18, 88]
]

print("Студенты старше 20 лет:")
for row in data:
    if row[1] > 20:
        print(f"- {row[0]} ({row[1]} лет)")

print("\nСтуденты с баллом выше 80:")
for row in data:
    if row[2] > 80:
        print(f"- {row[0]} (балл: {row[2]})")
```
