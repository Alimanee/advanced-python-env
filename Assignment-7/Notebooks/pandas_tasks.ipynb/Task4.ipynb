```python
print("Исследование данных")

data = [
    ["Али", 19, 85],
    ["Мария", 21, 92],
    ["Иван", 20, 78],
    ["Елена", 22, 95],
    ["Дмитрий", 18, 88]
]

print("Первые 3 строки:")
for i in range(3):
    print(f"{data[i][0]}, {data[i][1]} лет, балл: {data[i][2]}")

scores = [row[2] for row in data]
average_score = sum(scores) / len(scores)
print(f"\nСредний балл: {average_score:.2f}")

ages = [row[1] for row in data]
print(f"Средний возраст: {sum(ages)/len(ages):.1f} лет")
```
